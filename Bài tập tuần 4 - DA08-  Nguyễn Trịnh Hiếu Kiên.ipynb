```python
#Vẽ phương trình y = ax +b (cho phép gán giá trị khác nhauc của a và b)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Nhập giá trị của a và b 
a = int(input("Nhập giá trị của a: "))
b = int(input("Nhập giá trị của b: "))

# Tạo một mảng x từ -5 đến 5
x = np.arange(-5, 6)
# Tính giá trị tương ứng của y từ a và b
y = a*x + b
# Vẽ và hiển thị
plt.title(f"Biểu đồ của hàm :{a}x+{b}")
plt.plot(x, y)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv("C:\DA08\data-stock\excel_acb.csv")

# Vẽ biểu đồ giá Open và Close
plt.plot(data["<OpenFixed>"], label="Open",color="blue",linewidth=1)
plt.plot(data["<CloseFixed>"], label="Close", color="red",linewidth=1)
plt.axhline(y=data["<OpenFixed>"].mean(), color="gray", linestyle="--", linewidth=1)
# Đặt tên cho trục x và trục y
plt.xlabel("Ngày")
plt.ylabel("Giá chứng khoán")

# Đặt tiêu đề cho biểu đồ
plt.title("Biểu đồ giá Open và Close ACB")

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
#Vẽ biểu đồ cho nhiều mã chứng khoán khác nhau
import pandas as pd
import matplotlib.pyplot as plt

# Danh sách các mã chứng khoán cần vẽ biểu đồ
stock_codes = ["ACB", "DXG", "FLC", "FPT","VJC","VNM"]

# Vẽ biểu đồ cho mỗi mã chứng khoán
for code in stock_codes:
    # Đọc dữ liệu từ tệp CSV
    data = pd.read_csv(f"C:\DA08\data-stock\excel_{code}.csv")

    # Vẽ biểu đồ giá Open và Close
    plt.plot(data["<OpenFixed>"], label=f"{code} - Open")
    plt.plot(data["<CloseFixed>"], label=f"{code} - Close")

# Đặt tên cho trục x và trục y
plt.xlabel("Ngày")
plt.ylabel("Giá chứng khoán")

# Đặt tiêu đề cho biểu đồ
plt.title("Biểu đồ giá Open và Close của các mã chứng khoán")

# Hiển thị chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
# Nhập giá trị của a và b 
a = int(input("Nhập giá trị của a: "))
b = int(input("Nhập giá trị của b: "))
c = int(input("Nhập giá trị của c: "))
# Tạo một mảng x từ -5 đến 5
x = np.arange(-5, 6)
# Tính giá trị tương ứng của y từ a và b
y = a*x*x + b*x
# Vẽ và hiển thị
plt.title(f"Biểu đồ của hàm :{a}x^2+{b}x+{c}")
plt.plot(x, y)
plt.show()
```
